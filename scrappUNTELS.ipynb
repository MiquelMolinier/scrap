```python
import pickle
import requests
from bs4 import BeautifulSoup
```

```python
rev = "http://revistas.untels.edu.pe/index.php/files/issue/archive"
req_rev = requests.get(rev)
soup_rev = BeautifulSoup(req_rev.content, "lxml")
```

```python
vols = []
revi = soup_rev.find_all("div", attrs="issue_item")
for vol in revi:
    dictt = {}
    vol = vol.div
    tags = vol.find_all("a")[1:]
    spans = tags[0].find_all("span")
    title = f"{spans[0].string} {spans[1].string} {spans[2].string}"
    title += f' {tags[1].string.strip()}'
    archive = title.replace(" ","_").replace("&","")
    link_vol = tags[1]["href"]
    dictt["title"] = title
    dictt["link_vol"] = link_vol
    dictt["archive"] = archive
    vols.append(dictt)
vols
```

```python
for vol in vols:
    req_vol = requests.get(vol["link_vol"])
    soup_vol = BeautifulSoup(req_vol.content, "lxml")
    arts = soup_vol.find("div",attrs="section_content").find_all("article", attrs="article_summary")
    artt = []
    for art in arts:
        dictt = {}
        summary = art.find("div",attrs="article_summary_body")
        link_art = summary.h5.a["href"]
        authors = art.find("div","summary_meta").div.string
        authors = authors.replace("\n","").replace("\t","")
        authors = authors.split(",")
        title = summary.h5.a.string.replace("\n","").replace("\t","")
        dictt["link_art"] = link_art
        dictt["authors"] = authors
        dictt["title"] = title
        artt.append(dictt)
    vol["articles"] = artt
```

```python
vols
```

```python
for vol in vols:
    for art in vol["articles"]:
        req_art = requests.get(art["link_art"])
        soup_art = BeautifulSoup(req_art.content, "lxml")
        date = soup_art.find("div",attrs="main_entry col-md-4").p.string
        date = date.replace("\n","").replace("\t","").replace(" ","")
        t = date.find("Publicado")
        date = date[t+9:]
        art["date"] = date
        art["soup"] = soup_art
```

```python
with open("scrappUNTELS.pickle", "wb") as f:
    pickle.dump(vols, f)
```
