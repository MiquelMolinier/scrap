```python
import pickle
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
scrappUNMSMfisi = None
scrappULASALLE = None
scrappUNMSMcomp = None
scrappUNTELS = None
scrappURP = None
scrappULIMA = None
with open("scrappUNMSMfisi.pickle", "rb") as f:
    scrappUNMSMfisi = pickle.load(f)
with open("scrappUNMSMcomp.pickle", "rb") as f:
    scrappUNMSMcomp = pickle.load(f)
with open("scrappULASALLE.pickle", "rb") as f:
    scrappULASALLE = pickle.load(f)
with open("scrappUNTELS.pickle", "rb") as f:
    scrappUNTELS = pickle.load(f)
with open("scrappURP.pickle", "rb") as f:
    scrappURP = pickle.load(f)
with open("scrappULIMA.pickle", "rb") as f:
    scrappULIMA = pickle.load(f)
```

```python
for vol in scrappUNMSMfisi:
    vol["journal"] = "Revista de investigación de Sistemas e Informática"
    print(vol["title"])
print("-"*50)
for vol in scrappUNMSMcomp:
    vol["journal"] = "Revista peruana de computación y sistemas"
    print(vol["title"])
print("-"*50)
for vol in scrappULASALLE:
    vol["journal"] = "Innovación y Software"
    print(vol["title"])
print("-"*50)
for vol in scrappUNTELS:
    vol["journal"] = "BIOTECH & ENGINEERING"
    if vol["title"].count("02") == 2:
        vol["title"] = vol["title"].replace("2021", "(2021)").replace("02", "2", 1).replace("No", "Núm.").replace("Vol", "Vol.")
    else:
        vol["title"] = vol["title"].replace("2021", "(2021)").replace("01", "1", 1).replace("No", "Núm.").replace("Vol", "Vol.")
    print(vol["title"])
print("-"*50)
for vol in scrappURP:
    vol["journal"] = "Perfiles de Ingeniería"
    print(vol["title"])
for vol in scrappULIMA:
    vol["journal"] = "Interfases"
    print(vol["title"])
print("-"*50)
```

```python
scrapps = scrappUNMSMfisi + scrappULASALLE + scrappUNMSMcomp + scrappUNTELS + scrappURP + scrappULIMA
for vol in scrapps:
    title = vol["title"]
    vol_i = title.find("Vol.") + 4
    vol_d = title.find("Núm.")
    num_i = vol_d + 4
    num_d = title.find("(")
    volumen = title[vol_i+1: vol_d-1]
    num = title[num_i+1: num_d-1]
    year = title[num_d+1: num_d+5]
    vol["year"] = year
    vol["volumen"] = volumen
    vol["numero"] = num
```

```python
journal = []
volumen = []
numero = []
year = []
for vol in scrapps:
    journal.append(vol["journal"])
    volumen.append(vol["volumen"])
    numero.append(vol["numero"])
    year.append(vol["year"])
```

```python
df = pd.DataFrame({
    "id_revista": np.arange(len(scrapps)),
    "journal": journal,
    "volumen": volumen,
    "numero": numero,
    "year": year
})
writer = pd.ExcelWriter("journals.xlsx")
df.to_excel(writer, "Hoja de datos", index=False)
writer.save()
```

```python
arts = []
for i, vol in enumerate(scrapps):
    for art in vol["articles"]:
        art["id_revista"] = i
        arts.append(art)
```

```python
date = []
title = []
id_revista = []
for art in arts:
    date.append(art["date"])
    title.append(art["title"])
    id_revista.append(art["id_revista"])
```

```python
df = pd.DataFrame({
    "id_articulo": np.arange(len(arts)),
    "id_revista": id_revista,
    "date": date,
    "title": title,
})
writer = pd.ExcelWriter("articles.xlsx")
df.to_excel(writer, "Hoja de datos", index=False)
writer.save()
```
