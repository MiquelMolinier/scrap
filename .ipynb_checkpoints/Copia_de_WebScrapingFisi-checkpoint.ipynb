##Lista total de elementos que se van a almacenar

```python
#
lista_diccionarios_final = []
```

Aquí cargo los campos en forma de diccionario en una lista

##Se va a almacenar el nombre o titulo de la revista con el link de la direccion a la revista

```python
for i in range(0,len(listavolumenes)):
  dict_aux={}
  dict_aux['title_volumen']=str(listavolumenes[i].find('a',attrs='title').get_text().replace('\n','').replace('\t','')).strip()
  print(str(listavolumenes[i].find('a',attrs='title').get_text().replace('\n','').replace('\t','')).strip())
  dict_aux['link_volumen']=listavolumenes[i].a.get('href')
  print(listavolumenes[i].a.get('href'))
  lista_diccionarios_final.append(dict_aux)
  print("\n")
  #direccion_volumenes
```

```python
lista_diccionarios_final
```

Con esto se va trabajar  con la lista para agrupar todo los elementos

```python
lista_diccionarios_final[1]['link_volumen']
```

```python
n = len(lista_diccionarios_final)
for i in range(0, 1):
    consultt = requests.get(lista_diccionarios_final[i]["link_volumen"])
    soupt = BeautifulSoup(consultt.content, "lxml")
    section = soupt.find_all("div", attrs="section")[1].ul
    articles = section.find_all("li")
    print(articles[0])
    with open("articles.pickle", "wb") as f:
        pickle.dump(articles, f)
```

```python
with open("articles.pickle", "rb") as f:
    obj = pickle.load(f)
    
```

```python
lista_diccionarios_final_direcciones=[]

for i in range(0,len(lista_diccionarios_final)):
  identi=i
  #for i in range(0,len(lista_diccionarios_final)):
  #print(lista_diccionarios_final[i]['link_volumen'])

  #vamos a hacer un requests de las diversas paginas de los cada revista 
  consulta_aux=requests.get(lista_diccionarios_final[i]['link_volumen'])

  #sopa de cada requests
  sopa_aux=BeautifulSoup(consulta_aux.content , 'lxml')


  #Aqui se ve cada elemento de las secciones que son solo dos en todas las paginas
  #elemento_articulos_aux ---> es una lista de 2 elementos 
  #Lo que yo voy a utilizar es la segunda componentes .... 

  #Seccion es una lista de un solo elemento.
  elemento_articulos_aux=sopa_aux.find_all('div',attrs='sections')



  #La seccion es de dos elementos y por eso es lo que ahora vamos a usar el elemento 2
  seccion_aux_arti = elemento_articulos_aux[0].find_all('div',attrs='section')


  #esto es el contenedor que esta dentro de todo el li 
  #seccion_aux_arti[1]  ya con esto estoy usando el elemento 2 del array
  elementos_aux_doc= seccion_aux_arti[1].find_all('div',attrs='obj_article_summary')


  #detalle de cada componente

  for j in range(0,len(elementos_aux_doc)):
    dict_a={}
    dict_a['referencia_articulo']= elementos_aux_doc[j].a.get('href')
    dict_a['link_padron'] = lista_diccionarios_final[identi]['link_volumen']
    dict_a['title_volumen'] = lista_diccionarios_final[identi]['title_volumen']


    consu_art_aux=requests.get(dict_a['referencia_articulo'])
    s_consu_art=BeautifulSoup(consu_art_aux.content,'lxml')

    #titulo del articulo
    titulo_articulos_feed=s_consu_art.find('h1',attrs='page_title')
    dict_a['title_articulo']=titulo_articulos_feed.text.replace('\n','').replace('\t','')


    #Autores 
    autores=''
    autores_aux = s_consu_art.find('ul',attrs='authors').find_all('span',attrs='name')
    len(autores_aux)

    for k in range(0,len(autores_aux)):
      autores=autores + str(autores_aux[k].text.replace('\n','').replace('\t','')) + ' - '
    dict_a['autores']=autores


    #pdf--
    pdf_articulo = s_consu_art.find('a',attrs='obj_galley_link pdf').get('href')
    dict_a['pdf_articulo']=pdf_articulo


    #fechapublicacion
    fecha_publicacion = s_consu_art.find('section',attrs='sub_item').span.text
    dict_a['fecha_publicacion']=fecha_publicacion

    lista_diccionarios_final_direcciones.append(dict_a)
print(len(lista_diccionarios_final_direcciones))

```
