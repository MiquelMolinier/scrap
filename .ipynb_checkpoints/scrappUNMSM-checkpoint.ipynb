```python
import pickle
import requests
from bs4 import BeautifulSoup
```

```python
prev = "https://revistasinvestigacion.unmsm.edu.pe/"
req_prev = requests.get(prev)
soup_prev = BeautifulSoup(req_prev.content, "lxml")
```

```python
revcomp = "Revista peruana de computación y sistemas"
revfisi = "Revista de investigación de Sistemas e Informática"
revs = soup_prev.find("div", attrs="journals").ul.find_all("li", attrs="has_thumb")
revs
```

```python
link_comp = None
link_fisi = None
for rev in revs:
    linkt = rev.find("div", attrs="body").h3.a
    title_rev = linkt.string.replace("\n", "").replace("\t","")
    if title_rev == revcomp:
        link_comp = linkt["href"]
    if title_rev == revfisi:
        link_fisi = linkt["href"]
```

```python
print(link_comp)
print(link_fisi)
```

## Revista de computación

```python
req_volt_comp = requests.get(link_comp)
soup_volt_comp = BeautifulSoup(req_volt_comp.content, "lxml")
```

```python
temp = soup_volt_comp.find("section", attrs="current_issue").contents
tags_a = []
for tag in temp:
    if tag.name == "a":
        tags_a.append(tag)
new_link_comp = tags_a[-1]["href"]
new_link_comp
```

```python
req_vols_comp = requests.get(new_link_comp)
soup_vols_comp = BeautifulSoup(req_vols_comp.content, "lxml")
```

```python
soup_vols_comp = soup_vols_comp.find("ul", attrs="issues_archive").find_all("div", attrs="obj_issue_summary")
soup_vols_comp
```

```python
vols_comp = []
for vol in soup_vols_comp:
    dictt = {}
    link_vol = vol.h2.a
    title = link_vol.string
    title = title.replace("\n", "").replace("\t", "").strip()
    link_vol = link_vol["href"]
    archive = title.replace("ú", "u").replace(".", "").replace("(", "")
    archive = archive.replace(")", "").replace(" ", "_")
    dictt["title"] = title
    dictt["archive"] = archive
    dictt["link_vol"] = link_vol
    vols_comp.append(dictt)
vols_comp
```

```python
for vol_comp in vols_comp:
    req_vol_comp = requests.get(vol_comp["link_vol"])
    soup_vol_comp = BeautifulSoup(req_vol_comp.content, "lxml")
    sections = soup_vol_comp.find("div", attrs="sections")
    arts = sections.find_all("div", attrs="obj_article_summary")
    artt = []
    for art in arts:
        dictt = {}
        link_art = art.find("h3", attrs="title").a
        title = link_art.string
        title = title.replace("\t", "").replace("\n", "").strip()
        link_art = link_art["href"]
        authors = art.find("div", attrs="authors").string
        authors = authors.replace("\n", "").replace("\t", "").strip().split(",")
        dictt["link_art"] = link_art
        dictt["title"] = title
        dictt["authors"] = authors
        artt.append(dictt)
    vol_comp["articles"] = artt
```

```python
for vol_comp in vols_comp:
    for art_comp in vol_comp["articles"]:
        req_art = requests.get(art_comp["link_art"])
        soup_art = BeautifulSoup(req_art.content, "lxml")
        date = soup_art.find("div", attrs="item published").find("section", attrs="sub_item").div
        date = date.span.string
        art_comp["date"] = date
```

```python
print(vols_comp)
```

```python
with open("scrappUNMSMcomp.pickle", "wb") as f:
    pickle.dump(vols_comp, f)
```

```python
with open("scrappUNMSMcomp.pickle", "rb") as f:
    obj = pickle.load(f)
    print(obj)
```

## Revista de ingeniería de sistemas

```python
req_volt_fisi = requests.get(link_fisi)
soup_volt_fisi = BeautifulSoup(req_volt_fisi.content, "lxml")
```

```python
temp = soup_volt_fisi.find("section", attrs="current_issue").contents
tags_a = []
for tag in temp:
    if tag.name == "a":
        tags_a.append(tag)
new_link_fisi = tags_a[-1]["href"]
new_link_fisi
```

```python
req_vols_fisi = requests.get(new_link_fisi)
soup_vols_fisi = BeautifulSoup(req_vols_fisi.content, "lxml")
```

```python
soup_vols_fisi = soup_vols_fisi.find("ul", attrs="issues_archive").find_all("div", attrs="obj_issue_summary")
soup_vols_fisi
```

```python
vols_fisi = []
for vol in soup_vols_fisi:
    dictt = {}
    link_vol = vol.h2.a
    title = link_vol.string
    title = title.replace("\n", "").replace("\t", "").strip()
    link_vol = link_vol["href"]
    archive = title.replace("ú", "u").replace(".", "").replace("(", "")
    archive = archive.replace(")", "").replace(" ", "_")
    dictt["title"] = title
    dictt["archive"] = archive
    dictt["link_vol"] = link_vol
    vols_fisi.append(dictt)
vols_fisi
```

```python
vol_fisi = vols_fisi[0]
req_vol_fisi = requests.get(vol_fisi["link_vol"])
soup_vol_fisi = BeautifulSoup(req_vol_fisi.content, "lxml")
```

```python
for vol_fisi in vols_fisi:
    req_vol_fisi = requests.get(vol_fisi["link_vol"])
    soup_vol_fisi = BeautifulSoup(req_vol_fisi.content, "lxml")
    sections = soup_vol_fisi.find("div", attrs="sections")
    temp = sections.find_all("div", attrs="section")
    for section in temp:
        if not(section.find("h2") is None):
            if section.find("h2").string.replace("\t", "").replace("\n", "").strip()=="Artículos":
                break
    arts = section.find_all("div", attrs="obj_article_summary")
    artt = []
    for art in arts:
        dictt = {}
        link_art = art.find("h3", attrs="title").a
        title = link_art.string
        title = title.replace("\t", "").replace("\n", "").strip()
        link_art = link_art["href"]
        authors = art.find("div", attrs="authors").string
        authors = authors.replace("\n", "").replace("\t", "").strip().split(",")
        dictt["title"] = title
        dictt["link_art"] = link_art
        dictt["authors"] = authors
        artt.append(dictt)
    vol_fisi["articles"] = artt
```

```python
for vol_fisi in vols_fisi:
    for art_fisi in vol_fisi["articles"]:
        req_art = requests.get(art_fisi["link_art"])
        soup_art = BeautifulSoup(req_art.content, "lxml")
        date = soup_art.find("div", attrs="item published").find("section", attrs="sub_item").div
        date = date.span.string
        art_fisi["date"] = date
```

```python
print(vols_fisi)
```
