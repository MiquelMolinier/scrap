```python
import turicreate
import pandas as pd
import numpy as np
```

```python
articles = turicreate.SFrame.read_csv("articles.csv")
articles
```

```python
articles['word_count'] = turicreate.text_analytics.count_words(articles['title'])
articles
```

```python
article = articles[9]
article
```

```python
df = turicreate.SFrame.to_dataframe(articles)
df
```

```python
words = {}
for i in df.index:
    title = df["word_count"][i]
    for word in title.keys():
        count = words.setdefault(word, 0)
        words[word] += title[word]
words
```

```python
keys = list(words.keys())
values = list(words.values())
```

```python
id_article_f = []
id_word_f = []
for i in df.index:
    title = df["word_count"][i]
    for word in title:
        id_article_f.append(df["id_articulo"][i])
        id_word_f.append(keys.index(word))
```

```python
df = pd.DataFrame({
    'id_art_word': np.arange(len(id_article_f)),
    'id_articulo': id_article_f ,
    'id_word': id_word_f
})

writer = pd.ExcelWriter('art_word.xlsx')
df.to_excel(writer, 'art_word', index=False)
writer.save()
```

```python
df = pd.DataFrame({
    'id_word': np.arange(len(keys)),
    'word': keys ,
    'count': values
})
writer = pd.ExcelWriter('word.xlsx')
df.to_excel(writer, 'word', index=False)
writer.save()
```
